## 7. Find Out Number of Courses Per Subject"

## 9. Display The Count of Paid and Free Courses

```python
data.columns
```

```python
data['is_paid'].value_counts()
```

```python
sns.countplot(x='is_paid', data=data)
plt.xlabel('Course_type', fontsize=13)
plt.ylabel('Number of Free and Paid Courses', fontsize=13)
plt.xticks(rotation=65)
plt.xticks(ticks =[1, 0], labels =['Free', 'Paid'])
plt.show()
```

## 10. Which courses has more lectures (Free or Paid)?

```python
numeric_data = data.select_dtypes(include='number')
grouped_means = data.groupby(['is_paid'])[numeric_data.columns].mean()
print(grouped_means)
```
